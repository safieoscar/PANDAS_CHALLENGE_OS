* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#getting total number of players
players = len(purchase_data['SN'].value_counts())
#database formatting
total_players = pd.DataFrame(
    {"Total Players": [players]
    })
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#unique items
unique_items = len(purchase_data['Item Name'].value_counts())
#average price
average_price = round(purchase_data['Price'].mean(), 2)
#number of purchases
number_of_purchases = purchase_data['Purchase ID'].count()
#total revenue
total_revenue = purchase_data['Price'].sum()

#database formatting
purchasing_analysis_total = pd.DataFrame(
    {"Number of Unique Items": [unique_items], "Average Price": [average_price], 
     "Number of Purchases": [number_of_purchases], "Total Revenue": [total_revenue]
        })
pd.options.display.float_format = "${:,.2f}".format
purchasing_analysis_total
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#dropping duplicate customers
dropped_duplicates = purchase_data.drop_duplicates(subset='SN', keep="first")

#dropping out columns
dropped_columns = dropped_duplicates.drop(['Purchase ID', 'SN', 'Age', 'Item ID', 'Item Name', 'Price'], axis = 1)

#filtering out genders
male = dropped_columns.query('Gender == "Male"')
female = dropped_columns.query('Gender == "Female"')
other = dropped_columns.query('Gender == "Other / Non-Disclosed"') 

#counts per gender
male_count = len(male)
female_count = len(female)
other_count = len(other)

#percentage per gender
male_percent = round(male_count/players * 100, 2)
female_percent = round(female_count/players * 100, 2)
other_percent = round(other_count/players * 100, 2)

#dataframe formatting
gender_demographics = pd.DataFrame(
    {"Total Count": [male_count, female_count, other_count], 
     "Percentage of Players": [male_percent, female_percent, other_percent]
        })
gender_demographics.index = (["Male", "Female", "Other / Non-Disclosed"])
pd.options.display.float_format = "{:,.2f}%".format
gender_demographics


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins and their labels
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 99]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#adding existing players into the age bins
dropped_duplicates["Age Bin"] = pd.cut(dropped_duplicates["Age"], bins, labels= age_labels)

#adding bins into dataframe
age_group = dropped_duplicates.groupby("Age Bin")

#player count per bin
age_count = age_group["SN"].count()

#player percentage per bin
age_percent = round((age_count/players) * 100, 2)

#database formatting
age_demographics = pd.DataFrame(
    {"Total Count": age_count, "Percentage of Players": age_percent   
    })
pd.options.display.float_format = "{:,.2f}%".format
age_demographics.index.name = None
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#using data without dropped duplicates
purchase_data["Ages Bin"] = pd.cut(purchase_data["Age"], bins, labels= age_labels)

#adding bins into dataframe
undropped = purchase_data.groupby("Ages Bin")

#getting purchase count
purchase_count = undropped["Purchase ID"].count()

#average purchase price
average_purchase = round(undropped["Price"].mean(), 2)

#total purchase value
total_purchase = undropped["Price"].sum()

#average total purchase per person
avg_total_per_person = round(total_purchase/age_count, 2)

#dataframe formatting
purchasing_analysis_age = pd.DataFrame({
    "Purchase Count": purchase_count, "Average Purchase Price": average_purchase, "Total Purchase Value": total_purchase, 
    "Avg Total Purchase per Person": avg_total_per_person
})
pd.options.display.float_format = "${:,.2f}".format
purchasing_analysis_age.index.name = "Age Ranges"
purchasing_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#groupby dataframe by SN
spenders = purchase_data.groupby("SN")

#get purchase count per player
purchase_spenders = spenders["Purchase ID"].count()

#average spending by player
average_spending = round(spenders["Price"].mean(), 2)

#total purchase value
total_purchase_value = spenders["Price"].sum()

#dataframe formatting
top_spenders = pd.DataFrame({
    "Purchase Count": purchase_spenders, "Average Purchase Price": average_spending, 
    "Total Purchase Value": total_purchase_value
        })
pd.options.display.float_format = "${:,.2f}".format
top_spenders.index.name = "SN"

#sorting from greatest to least spending player
sorted_top_spenders = top_spenders.sort_values(["Total Purchase Value"], ascending = False).head()
sorted_top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group dataframe by Item ID and Item Name
items = purchase_data[["Item ID", "Item Name", "Price"]]

items_group = items.groupby(["Item ID", "Item Name"])

#get purchase count per Item
purchase_count_popular = items_group["Price"].count()

#total purchase value
total_purchase_value_top = items_group["Price"].sum()

#get item prices
top_item_price = round(total_purchase_value_top/purchase_count_popular, 2)

#dataframe formatting
most_popular_items = pd.DataFrame({
    "Purchase Count": purchase_count_popular, "Item Price": top_item_price, "Total Purchase Value": total_purchase_value_top
})

pd.options.display.float_format = "${:,.2f}".format

#sorting by purchase count
sorted_most_popular_items = most_popular_items.sort_values(["Purchase Count"], ascending = False).head()
sorted_most_popular_items
```
